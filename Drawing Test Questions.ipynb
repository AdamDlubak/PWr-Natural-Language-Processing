Import Libraries
--------------

```python
import pandas as pd
import json
import random
```

Parameters
--------------

```python
dataset_file = "czywiesz-eva-I-250.csv"
json_file = "questions_to_test.json"
number_of_questions = 50
```

Prepare Dataset
-------------

```python
dataset = pd.read_csv(dataset_file, header=None, sep=";")
display(dataset.head(5))

dataset = dataset.drop(columns=[2, 3])
dataset.rename(columns={1: 'Question', 0: 'Question_ID'}, inplace=True)
dataset['Dataset_ID'] = dataset.index
display(dataset.head(5))
```

Draw Quesstions
-------------

```python
questions_index = random.sample(range(0, dataset.shape[0] - 1 ), number_of_questions)
questions_to_test = pd.DataFrame(columns=["Question_ID", "Question", "Dataset_ID"])
for idx in questions_index:
    questions_to_test = questions_to_test.append(dataset.iloc[[idx]])
    
display(questions_to_test)
```

Save questions to JSON file
-----------

```python
questions_data = questions_to_test.to_json(orient='split')

with open(json_file, 'w') as outfile:
    json.dump(questions_data, outfile)
```
